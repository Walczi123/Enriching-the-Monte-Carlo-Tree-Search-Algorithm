```python
import pandas as pd
```

```python
df = pd.read_csv(FILE_PATH, sep=SEPARATHOR, header=None)

df.columns = ['game_type','player1','player2','winner','seed','game_time']

df = df[df['game_type']=='hex']

df['time_min']=df['game_time']/(60)
df['time_h']=df['game_time']/(60*60)
df['time_day']=df['game_time']/(24*60*60)
```

```python
df.groupby(by=['player1'], as_index=True).agg({'game_time':'count',
'time_min': 'mean',
'time_h': 'mean',
'time_day': 'mean'})
```

```python
df.groupby(by=['player2'], as_index=True).agg({'game_time':'count',
'time_min': 'mean',
'time_h': 'mean',
'time_day': 'mean'})
```
